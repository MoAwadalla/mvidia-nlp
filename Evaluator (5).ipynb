### Modify Sentences to literal meaning before translation

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, AutoProcessor
```

```python
phi_model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
phi_tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
phi_pipe = pipeline(
    "text-generation",
    model=phi_model,
    tokenizer=phi_tokenizer,
  )
generation_args = {
  "max_new_tokens": 500,
  "return_full_text": False,
  "do_sample": False
}
```

```python
def modify_idiom_phi(idiom):
  messages = [
    {"role" : "user", "content": f"If the Urdu sentence doesn't have any figurative or idioms, then output the Urdu sentence exactly as is. If the sentence has figurative language or idioms, output an Urdu sentence of the same meaning without any idioms or figurative language in Urdu. Do not attempt to explain the meaning, ONLY output the modified sentence in Urdu.  Sentence: {idiom}"},
  ]
  phi_output = phi_pipe(messages, **generation_args)
  modified_idiom = phi_output[0]['generated_text']

  print(f'Original sentence: {idiom}')
  print(f'Modified sentence: {modified_idiom}')
  return modified_idiom
```

```python
def modify_idiom(sentence):
  messages = [
    {"role" : "user", "content": f"If the Urdu sentence doesn't have any figurative or idioms, then output the Urdu sentence exactly as is. If the sentence has figurative language or idioms, output an Urdu sentence of the same meaning without any idioms or figurative language in Urdu. Do not attempt to explain the meaning, ONLY output the modified sentence in Urdu.  Sentence: {sentence}"},
  ]

  prompt = modification_pipeline.tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
  )

  # Define end-of-sentence terminators
  terminators = [
    modification_pipeline.tokenizer.eos_token_id,
    modification_pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  results = modification_pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9
  )

  modified_sentence = results[0]["generated_text"][len(prompt):]

  print(f'Original sentence: {sentence}')
  print(f'Modified sentence: {modified_sentence}')
  return modified_sentence
```

```python
def calculate_bleu(translated_idioms):
  references = []
  candidates = []
  for _, translated_text, expected_english in translated_idioms:
    # Tokenize the expected and generated translations
    reference = nltk.word_tokenize(expected_english.lower())
    candidate = nltk.word_tokenize(translated_text.lower())

    # For corpus BLEU scores, references are lists of lists, candidates are lists
    references.append([reference])  # reference needs to be a list of lists
    candidates.append(candidate)

  # Calculate the BLEU score
  bleu_score = corpus_bleu(references, candidates)
  return bleu_score

def calculate_exact_match_score(translated_idioms):
  exact_matches = sum(1 for _, translated, expected in translated_idioms if translated.strip().lower() == expected.strip().lower())
  return exact_matches / len(translated_idioms)

def weighted_score(bleu_score, exact_match_score, bleu_weight=0.5, exact_match_weight=0.5):
  return (bleu_score * bleu_weight) + (exact_match_score * exact_match_weight)
```

use difference between embeddings to calculate accuracy of translations

```python
import torch
from transformers import AutoTokenizer, AutoModel
from scipy.spatial.distance import cosine

# Load sentence embedding model
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
model = AutoModel.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
```

```python
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=256)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(1)

def calculate_similarity(embedding1, embedding2):
    # Detach embeddings from the computation graph, ensure they are 1-D, and convert to numpy arrays
    emb1 = embedding1.detach().squeeze().numpy()
    emb2 = embedding2.detach().squeeze().numpy()
    return 1 - cosine(emb1, emb2)

def evaluate_translations(translations):
    total_cosine_similarity = 0
    total_exact_match = 0
    num_samples = len(translations)

    for urdu_idiom, model_output, correct_translation in translations:
        # Get embeddings
        model_output_embedding = get_embedding(model_output)
        correct_translation_embedding = get_embedding(correct_translation)

        # Calculate cosine similarity
        cosine_sim = calculate_similarity(model_output_embedding, correct_translation_embedding)
        total_cosine_similarity += cosine_sim

        # Calculate exact match
        exact_match = 1 if model_output.strip().lower() == correct_translation.strip().lower() else 0
        total_exact_match += exact_match

    # Calculate weighted average (adjust weights as needed)
    average_cosine_similarity = total_cosine_similarity / num_samples
    average_exact_match = total_exact_match / num_samples
    weighted_average = 0.7 * average_cosine_similarity + 0.3 * average_exact_match

    return {
        "average_cosine_similarity": average_cosine_similarity,
        "average_exact_match": average_exact_match,
        "weighted_average": weighted_average
    }
```

```python
results_modified_SM4T = evaluate_translations(translated_modified_idioms_GPT4)
print("Average Cosine Similarity:", results_modified_SM4T["average_cosine_similarity"])
print("Average Exact Match Score:", results_modified_SM4T["average_exact_match"])
print("Weighted Average Score:", results_modified_SM4T["weighted_average"])
```

### OpenAI Embedding Cosine Similarity Calculation


```python
import openai
import numpy as np
from scipy.spatial.distance import cosine

# Set up OpenAI key
openai.api_key = ''

def get_embeddings(texts):
    """
    Returns embeddings for a list of texts using OpenAI's embedding model.
    Handles splitting the input into chunks if it exceeds the 2048 limit.
    """
    all_embeddings = []
    chunk_size = 2048

    for i in range(0, len(texts), chunk_size):
        chunk = texts[i:i + chunk_size]
        response = openai.embeddings.create(
        model="text-embedding-3-large",
        input=chunk
        )
        all_embeddings.extend([np.array(item.embedding) for item in response.data])

    return all_embeddings

def cosine_similarity(embedding1, embedding2):
    """
    Calculates the cosine similarity between two embeddings.
    """
    return 1 - cosine(embedding1, embedding2)

def evaluate_translations(translations):
    """
    Evaluates translations based on cosine similarity of embeddings and exact match.
    Expects translations as a list of tuples (source, model_output, ground_truth).
    """
    urdu_idioms, model_outputs, correct_translations = zip(*translations)
    embeddings_model_outputs = get_embeddings(model_outputs)
    embeddings_correct_translations = get_embeddings(correct_translations)

    average_cosine_similarity = 0
    exact_match_count = 0
    total = len(translations)

    results = []
    for (source, model_output, correct_translation), emb_mo, emb_gt in zip(translations, embeddings_model_outputs, embeddings_correct_translations):
        similarity = cosine_similarity(emb_mo, emb_gt)
        average_cosine_similarity += similarity
        exact_match = int(model_output.strip().lower() == correct_translation.strip().lower())
        exact_match_count += exact_match
        results.append((source, model_output, correct_translation, similarity, exact_match))

    average_cosine_similarity /= total
    exact_match_score = exact_match_count / total
    weighted_average = 0.7 * average_cosine_similarity + 0.3 * exact_match_score

    return {
        "results": results,
        "average_cosine_similarity": average_cosine_similarity,
        "average_exact_match": exact_match_score,
        "weighted_average": weighted_average
    }
```

### Using BERT Embedding Space

```python
import torch
from transformers import BertTokenizer, BertModel
import numpy as np
from scipy.spatial.distance import cosine

# Function to get the BERT embeddings
def get_bert_embeddings(text_list):
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')

    embeddings = []
    with torch.no_grad():
        for text in text_list:
            inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
            outputs = model(**inputs)
            cls_embedding = outputs.last_hidden_state[:, 0, :].numpy()
            embeddings.append(cls_embedding)

    return np.vstack(embeddings)

def calculate_similarity(embedding1, embedding2):
    """
    Calculates the cosine similarity between two embeddings.
    """
    return 1 - cosine(embedding1.flatten(), embedding2.flatten())

def evaluate_translations_bert(translations):
    """
    Evaluates translations based on cosine similarity of embeddings and exact match.
    Expects translations as a list of tuples (source, model_output, ground_truth).
    """
    urdu_idioms, model_outputs, correct_translations = zip(*translations)

    # Ensure inputs are valid strings
    model_outputs = [mo if mo.strip() else " " for mo in model_outputs]
    correct_translations = [ct if ct.strip() else " " for ct in correct_translations]

    embeddings_model_outputs = get_bert_embeddings(model_outputs)
    embeddings_correct_translations = get_bert_embeddings(correct_translations)

    average_cosine_similarity = 0
    exact_match_count = 0
    total = len(translations)

    results = []

    for (source, model_output, correct_translation), emb_mo, emb_gt in zip(translations, embeddings_model_outputs, embeddings_correct_translations):
        similarity = calculate_similarity(emb_mo, emb_gt)
        average_cosine_similarity += similarity
        exact_match = int(model_output.strip().lower() == correct_translation.strip().lower())
        exact_match_count += exact_match
        results.append((source, model_output, correct_translation, similarity, exact_match))

    average_cosine_similarity /= total
    exact_match_score = exact_match_count / total
    weighted_average = 0.7 * average_cosine_similarity + 0.3 * exact_match_score

    return {
        "results": results,
        "average_cosine_similarity": average_cosine_similarity,
        "average_exact_match": exact_match_score,
        "weighted_average": weighted_average
    }
```

```python
import cohere
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Initialize the Cohere client with your API key
api_key = ""
co = cohere.Client(api_key)
```

```python
# Function to get the Cohere embeddings
def get_cohere_embeddings(text_list):
    response = co.embed(model='embed-english-v3.0', texts=text_list, input_type='search_query')
    return response.embeddings

def calculate_similarity(embedding1, embedding2):
    """
    Calculates the cosine similarity between two embeddings.
    """
    return cosine_similarity([embedding1], [embedding2]).flatten()[0]

def evaluate_translations_cohere(translations):
    """
    Evaluates translations based on cosine similarity of embeddings and exact match.
    Expects translations as a list of tuples (source, model_output, ground_truth).
    """
    urdu_idioms, model_outputs, correct_translations = zip(*translations)

    # Ensure inputs are valid strings
    model_outputs = [mo if mo.strip() else " " for mo in model_outputs]
    correct_translations = [ct if ct.strip() else " " for ct in correct_translations]

    embeddings_model_outputs = get_cohere_embeddings(model_outputs)
    embeddings_correct_translations = get_cohere_embeddings(correct_translations)

    average_cosine_similarity = 0
    exact_match_count = 0
    total = len(translations)

    results = []

    for (source, model_output, correct_translation), emb_mo, emb_gt in zip(translations, embeddings_model_outputs, embeddings_correct_translations):
        similarity = calculate_similarity(emb_mo, emb_gt)
        average_cosine_similarity += similarity
        exact_match = int(model_output.strip().lower() == correct_translation.strip().lower())
        exact_match_count += exact_match
        results.append((source, model_output, correct_translation, similarity, exact_match))

    average_cosine_similarity /= total
    exact_match_score = exact_match_count / total
    weighted_average = 0.7 * average_cosine_similarity + 0.3 * exact_match_score

    return {
        "results": results,
        "average_cosine_similarity": average_cosine_similarity,
        "average_exact_match": exact_match_score,
        "weighted_average": weighted_average
    }
```

```python
def get_similarity_scores(phrases, ground_truth):
    """
    Returns individual cosine similarity scores for each phrase against the ground truth.

    :param phrases: List of phrases to compare.
    :param ground_truth: The ground truth phrase to compare against.
    :return: List of tuples with phrase and its similarity score.
    """
    # Ensure inputs are valid strings
    phrases = [phrase if phrase.strip() else " " for phrase in phrases]
    ground_truth = ground_truth if ground_truth.strip() else " "

    # Get embeddings for phrases and ground truth
    embeddings_phrases = get_cohere_embeddings(phrases)
    embedding_ground_truth = get_cohere_embeddings([ground_truth])[0]

    similarity_scores = []

    for phrase, embedding in zip(phrases, embeddings_phrases):
        similarity = calculate_similarity(embedding, embedding_ground_truth)
        similarity_scores.append((phrase, similarity))

    return similarity_scores

# Example usage
phrases = ["The scalded cat fears even cold water.", "A shy person beats around the bush in anger.", "A frustrated cat scratches the post."]
gt = "A confused person often makes things worse"

phrases2 = ["The path to the village", "The road to nowhere", "The way to the village"]
gt2 = "The path to wisdom"

similarity_scores = get_similarity_scores(phrases, gt)
similarity_scores2 = get_similarity_scores(phrases2, gt2)
print(gt)
print(similarity_scores)
print(gt2)
print(similarity_scores2)
```
